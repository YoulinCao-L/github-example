## Capstone Project The Battle of Neighborhoods
### Author: Youlin Cao

### 1. Problem Description
In this project we will try to find an optimal location for a restaurant. Specifically, this report will be targeted to stakeholders interested in opening an Coffee Shop in Stockholm, Sweden.

Since there are lots of restaurants in Stockholm we will try to detect locations that are not already crowded with restaurants. We are also interested in areas with no coffee shop in vicinity. 

We will use our data science powers to generate a few most promissing neighborhoods based on this criteria. Advantages of each area will then be clearly expressed so that best possible final location can be chosen by stakeholders.
